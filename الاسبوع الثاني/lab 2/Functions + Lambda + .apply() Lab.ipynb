```python
import pandas as pd

df = pd.DataFrame({
    "student": ["Ali", "Zeyad", "Omar", "3Bod", "W509_", "iZDX_7"],
    "Top": [90, 78, 65, 88, None, 55],
    "WOW": [85, 80, None, 92, 75, 60],
    "ZDX": [88, None, 70, 90, 80, 58]
})

df[["Top", "WOW", "ZDX"]] = df[["Top", "WOW", "ZDX"]].apply(lambda c: c.fillna(c.mean()))
df["average"] = df[["Top", "WOW", "ZDX"]].apply(lambda r: r.mean(), axis=1)
df["passed"] = df["average"].apply(lambda x: x >= 70)

df.to_csv("students_results.csv", index=False)
```
